```python
from bioservices import BioMart

def uniprot_to_ensembl(uniprot_list):
    s = BioMart()
    s.new_query()

    # Specify the dataset to use
    s.add_dataset_to_xml("hsapiens_gene_ensembl")

    # Add the attributes you're interested in retrieving
    s.add_attribute_to_xml("ensembl_peptide_id")
    s.add_attribute_to_xml("uniprotswissprot")

    # Add a filter for the UniProt IDs
    s.add_filter_to_xml("uniprotswissprot", ",".join(uniprot_list))

    # Get the results
    results = s.query()

    # Parsing the results
    mappings = {}
    for line in results.split("\n"):
        if line:
            ens_id, uniprot_id = line.split("\t")
            mappings[uniprot_id] = ens_id

    return mappings

#uniprot_ids = ["P31749", "Q9Y618", ...]  # Replace with your list of UniProt IDs
# conversion_result = uniprot_to_ensembl(uniprot_ids)
# print(conversion_result)
```

```python
import requests

def uniprot_to_ensp_direct(uniprot_ids):
    # Construct the XML query
    xml_query = f"""
    <?xml version="1.0" encoding="UTF-8"?>
    <!DOCTYPE Query>
    <Query virtualSchemaName = "default" formatter = "TSV" header = "0" uniqueRows = "0" count = "" datasetConfigVersion = "0.6" >
        <Dataset name = "hsapiens_gene_ensembl" interface = "default" >
            <Filter name = "uniprotswissprot" value = "{','.join(uniprot_ids)}"/>
            <Attribute name = "ensembl_peptide_id" />
        </Dataset>
    </Query>
    """

    # Make the POST request to BioMart
    response = requests.post("http://www.ensembl.org/biomart/martservice?query=", data=xml_query)
    response.raise_for_status()  # Raise an error for bad responses

    # Return the list of ENSP IDs
    return response.text.strip().split("\n")[1:]

# Test
#uniprot_ids = ["P31749", "O15111"]  # Replace this with your list of UniProt IDs
# ensp_ids = uniprot_to_ensp_direct(uniprot_ids)
# print(ensp_ids)
```

```python
projections, timing = {}, {}
for name, transformer in embeddings.items():
    if name.startswith("Linear Discriminant Analysis"):
        data = matrix.copy()
        data.flat[:: matrix.shape[1] + 1] += 0.01  # Make X invertible
    else:
        data = matrix

    print(f"Computing {name}...")
    start_time = time()
    projections[name] = transformer.fit_transform(data)
    timing[name] = time() - start_time
```

```python
# based on this distribution create a set of 8 bins with about equal degree membership

import networkx as nx

# Create a sample graph
#G = nx.erdos_renyi_graph(100, 0.05)

# Get degrees and sort by degree
sorted_nodes_by_degree = sorted(G.degree(), key=lambda x: x[1])

# Determine the size of each bin
bin_size = len(sorted_nodes_by_degree) // 8

# Create the bins
bins = []
for i in range(8):
    start_index = i * bin_size
    end_index = (i + 1) * bin_size if i != 7 else len(sorted_nodes_by_degree)  # To handle the remainder in the last bin
    bins.append(sorted_nodes_by_degree[start_index:end_index])

# Print the bins with their range of degrees
for b in bins:
    print(f"Degree Range: {b[0][1]}-{b[-1][1]}, Number of nodes: {len(b)}")
```

```python

# Determine the size of each bin
bin_size = len(sorted_nodes_by_degree2) // 8

# Create the bins
bins = []
for i in range(8):
    start_index = i * bin_size
    end_index = (i + 1) * bin_size if i != 7 else len(sorted_nodes_by_degree2)  # To handle the remainder in the last bin
    # start_index = round_to_nearest_50(i * bin_size)
    # end_index = round_to_nearest_50( (i + 1) * bin_size) if i != 7 else round_to_nearest_50(len(sorted_nodes_by_degree2))  # To handle the remainder in the last bin
    bins.append(sorted_nodes_by_degree2[(start_index):(end_index)]) #round_to_nearest_50

# Round to nearest 50
def round_to_nearest_50(n):
    return round(n / 50.0) * 50

# Print the bins with their range of degrees
for b in bins:
    lower_bound = round_to_nearest_50(b[0][1]) #b[0][1] #round_to_nearest_50(b[0][1])
    upper_bound = round_to_nearest_50(b[-1][1]) #b[-1][1] #round_to_nearest_50(b[-1][1])
    # lower_bound = b[0][1] #b[0][1] #round_to_nearest_50(b[0][1])
    # upper_bound = b[-1][1] #b[-1][1] #round_to_nearest_50(b[-1][1])
    print(f"Degree Range: {lower_bound}-{upper_bound}, Number of nodes: {len(b)}")
```
